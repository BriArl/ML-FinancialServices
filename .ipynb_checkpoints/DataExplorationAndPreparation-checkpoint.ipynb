```python
# ID columns used to identify records do not have much predictive power, so I'm dropping them from the data
# 

transaction_data = transaction_data.drop(labels= ['index', 'nameOrig', 'nameDest'], axis =1)
transaction_data.sample(5)

# This leaves me with a subset of columns which brings me more closer to predictive power
```

```python
# This will display all the transactions that have been tagged as fraud
# results show me a highly skewed dataset. Of the about 192k records, only 7.5k been tagged as fraud.

transaction_data['isFraud'].value_counts()
```

```python
# The same data can be extracted using countplot 

plt.figure(figsize=(12, 8))
sns.countplot(x='isFraud', data = transaction_data);

# When looking at the generated bar plot, it will show very few transactionsare actually fraudulent
```

```python
# Displays how much transacitions have been flagged as fraud 

transaction_data['isFlaggedFraud'].value_counts()
```

```python
# Datasets flagged as fraud use a simple heuristic, see the size of the transactions involved and tag only very large transactions as fraud.
# This dose not mean that these transactions are actually fraudulent (they have just been tagged as such using heuristic).

plt.figure(figsize= (12, 8))
sns.countplot(x='isFlaggedFraud', data = transaction_data);
```

```python
# Diffrent visualization of the tagged fraud transactions
# This will set up a categorical plot, which on the x-axis will have diffrent types

plt.figure(figsize=(12, 8))

sns.catplot(x = 'type', y = 'amount', estimator= sum,
            hue = 'isFraud', col = 'isFlaggedFraud',
            data = transaction_data);

# Orange are actually fraudulent transactions, it will display all of the fraudulent
# transactions are either CASH_OUT or TRANSFER type transactions. But only very few 
# transactions have been flagged as fraudulent, and of those are the TRANSFER type.
```

```python
# Within this dataset there is a column called step, it maps a unit of time in the real world. 
# One step corresponds to 1 hour of time. In this data set there is a total of 744 steps.

plt.figure(figsize=(12, 8))

plt.ylim([0, 8000])
sns.histplot(transaction_data['step'], kde= True);

# This simulation has been run over a period of 30 days.
```

```python
# Can alocate more intresting patterns if I split the  step column to be displayed in a 24hr duration
# Where every 24 hrs makes up a day- A modulo operation where we perform modulo 24 will give us a daily value of step

transaction_data['step'] = transaction_data['step'] % 24
transaction_data.head()
```

```python
# Now that step duration has been altered, I will display a line plot of step

plt.figure(figsize=(12, 8))
sns.lineplot(x= 'step', y='amount', hue = 'type', ci = None,
            estimator= 'mean', data = transaction_data);

# This line plot tracks the average amount of the transaction at diffrent periods during the day.
# All the fraudulent transactions in this dataset are either CASH_OUT or TRANSFER transactions they peak early in the day
```

```python
# Based on the line plot, I'm going to see if that means fraudulent transactions occur early in the day
# Plotting step on the x-axis and having a distribution plot for is fraud and is not fraud
sns.displot(data = transaction_data, x = 'step', col = 'isFraud');

# In the graph results the column on the left displays the non-fraudulent transactions
# Can see these transactions are kind of spread out throughout the day.
```

```python
#This graph makes it eveident that fraudulent transactions are either CASH_OUT or TRANSFER type
# This confirms that I can focus on there two transaction types 

plt.figure(figsize= (12, 8))
sns.countplot(x ='type', hue = 'isFraud', data = transaction_data)
```

```python
# Can filter out other transaction types from the dataset

transaction_data = transaction_data.loc[(transaction_data.type =='TRANSFER') | \
                                        (transaction_data.type == 'CASH_OUT')]
transaction_data.shape

# This leaves me with around 88k recs.
# Now I can focus on this recs. as I feel the other transaction types are unlikely to be fraud
# This is a heuristic whe im using based on the data 
```

```python
# The transaction type in the dataset is a categorical variable 
# Going to rep this variable using one-hot encoding, will do so using the pd.get_dummies function

transaction_data = pd.concat([transaction_data,
                            pd.get_dummies(transaction_data['type'],
                                            prefix = 'type', drop_first = True)],
                            axis =1)
transaction_data.head()

# Using the pd.get_dummies function on the type column, true will drop that column we dont need
# Looking at the chart now I have a encoded column call type_TRANSFER, this will have a value 0 that corresponds to CASH_OUT transactions and the value 1 corresonds to TRANSFER transactions.
```

```python
# Now that the columns arenumerically encoding the transaction type, I will drop isFlagged (as it's not very predictive)

transaction_data = transaction_data.drop(labels=['type', 'isFlaggedFraud'], axis =1)
transaction_data.head()
```

```python
# Data set is still slightly skewed 

transaction_data['isFraud'].value_counts()
```

```python
# Will now apply some feature engineering to create two new columns from the orginal data
# These columns, I feel, have features that will have predictive power for fraudulent transactions
# The first engineered column tracks discrepancy in the orgin balance account
# The second column tracks the discrpency in the destination account 

transaction_data['origBalanceDiscrepancy'] = \
    transaction_data.newbalanceOrig + transaction_data.amount - transaction_data.oldbalanceOrg

transaction_data['destBalanceDiscrepancy'] = \
    transaction_data.oldbalanceDest + transaction_data.amount - transaction_data.newbalanceDest
```

```python
# I ran a categorical plot of this discrepency based on fraudulent and non-fraudulent transactions

sns.catplot (x = 'isFraud', y = 'origBalanceDiscrepancy', estimator = sum,
            hue = 'type_TRANSFER', data = transaction_data, aspect =2) 

# After running my graph its evident that this feature may not add any interesting patterns to my ML model
```

```python
# I decided to track my destination balance in the same way based on fraudulent and non fraudulent (and based on type of transfer)
# in this categoricalstrip plot a pattern clear
# Discrepency in the destination accont is very high for fraudulent transactions
sns.catplot(x = 'isFraud', y = 'destBalanceDiscrepancy', estimator = sum,
            hue = 'type_TRANSFER', data = transaction_data, aspect = 2)
```

```python
# Now that I have all of the features that I need, I'm going to write this out to a CSV file 

transaction_data.to_csv('datasets/preprocessed_transaction_data.csv', index=False)
```
